## Markable extraction

###### Extract all named entities using the information from the 11th column

```python
result = pd.DataFrame(columns = ('doc_id','sentence','word_number','word','tagging','?','lemma','fourth*','FEATS','head','gramFct','co_num'))
co_num = 0
i=0
while i < len(df2):
    #find id a token belongs to two named entities
    if df2.iloc[i]['fourth*'].find('|'):
        NE_tags = df2.iloc[i]['fourth*'].split('|')
    else:
        NE_tags = df2.iloc[i]['fourth*']
        
    #if !='*' means that it's a start of a named entity 
    #or a named entity with only one word in it 
    #or the end of a named entity
    if df2.iloc[i]['fourth*']!='*':
        list_temp = []
        for tag in range(len(NE_tags)):  
            current = i
            start = df2.iloc[i]
            start_sync = df_sync_2.iloc[i]
            NE_tag = NE_tags[tag]
            
            #if it's a named entity with only one word in it 
            if NE_tag.startswith('(') and NE_tag.endswith(')'):
                #put the record in the result
                result = result.append(pd.DataFrame({'doc_id':[start['doc_id']],'sentence':[start['sentence']],'word_number':[start['word_number']],'word':[start['word']],'tagging':[start['tagging']],'?':[start['?']],'lemma':[start['lemma']],'fourth*':[start['fourth*']],'FEATS':[start_sync['FEATS']],'head':[start_sync['head']],'gramFct':[start_sync['gramFct']],'co_num':[co_num]}),sort=False)
                co_num = co_num+1
                length = 1
                
            #if it's a start of a named entity 
            elif NE_tag.startswith('(') and NE_tag.endswith('*') :
                length = 0
                #put the start record in the record and store it in a temporary list
                record = pd.DataFrame({'doc_id':[start['doc_id']],'sentence':[start['sentence']],'word_number':[start['word_number']],'word':[start['word']],'tagging':[start['tagging']],'?':[start['?']],'lemma':[start['lemma']],'fourth*':[start['fourth*']],'FEATS':[start_sync['FEATS']],'head':[start_sync['head']],'gramFct':[start_sync['gramFct']],'co_num':[co_num]})
                list_temp.append(record)
                while True:
                    i=i+1
                    start = df2.iloc[i]
                    start_sync = df_sync_2.iloc[i]
                    NE_tag = df2.iloc[i]['fourth*']
                    #unless run into '*)' which means it's the end of the named entity
                    if NE_tag.endswith('*)'):
                        #store the end in the record and the temporary list
                        record = pd.DataFrame({'doc_id':[start['doc_id']],'sentence':[start['sentence']],'word_number':[start['word_number']],'word':[start['word']],'tagging':[start['tagging']],'?':[start['?']],'lemma':[start['lemma']],'fourth*':[start['fourth*']],'FEATS':[start_sync['FEATS']],'head':[start_sync['head']],'gramFct':[start_sync['gramFct']],'co_num':[co_num]})
                        list_temp.append(record)
                        if length <= len(list_temp):
                            length = len(list_temp)
                        #store every dataframe in the temporary list to result
                        for temp in list_temp:
                            result = result.append(temp,sort = False)
                            co_num = co_num+1
                            i=i+1
                            i = current
                        break
                    #put every word in the middle of '(*' and '*)' in the record and store it in the temporary list
                    elif NE_tag == '*':
                        record = pd.DataFrame({'doc_id':[start['doc_id']],'sentence':[start['sentence']],'word_number':[start['word_number']],'word':[start['word']],'tagging':[start['tagging']],'?':[start['?']],'lemma':[start['lemma']],'fourth*':[start['fourth*']],'FEATS':[start_sync['FEATS']],'head':[start_sync['head']],'gramFct':[start_sync['gramFct']],'co_num':[co_num]})
                        list_temp.append(record)
        i=i+length
    else: i=i+1
            
```
